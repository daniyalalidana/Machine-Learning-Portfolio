## 1. Project Objective
The goal of this project is to build a Machine Learning model that predicts **Sales** based on advertising budgets for **TV, Radio, and Newspaper**.

**Business Question:**
> "Which marketing channel contributes the most to sales, and how can we optimize our budget?"

## 2. Methodology
1.  **Data Analysis:** Explore the relationship between ad spend and sales.
2.  **Preprocessing:** Split data into training and testing sets.
3.  **Model Building:** Train a Linear Regression model.
4.  **Evaluation:** Check the model's accuracy using MAE (Mean Absolute Error).

```python
#import Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn .model_selection import train_test_split
```

```python
#load data
url = "https://raw.githubusercontent.com/justmarkham/scikit-learn-videos/master/data/Advertising.csv"
data = pd.read_csv(url)

df = pd.DataFrame(data)
df.head()
```

## 3. Data Exploration & Analysis
We analyze the **Correlation Matrix** to see which features (TV, Radio, Newspaper) have the strongest relationship with Sales.

* **1.0** = Perfect Positive Correlation (As X goes up, Sales go up).
* **0.0** = No Correlation (Useless feature).

```python
# Correlation Analysis
print(df.corr()['Sales'].sort_values(ascending=False))

# Visualization
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## 4. Model Training
We split the data:
* **80% for Training:** The model learns from this.
* **20% for Testing:** We use this to test accuracy later.

```python
X = df[["TV", "Radio", "Newspaper"]]
y = df['Sales']

# Split Data (80% Train, 20% Test)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=54)
```

```python
# Train the Model
model = LinearRegression()
model.fit(X_train, y_train)

print("Model Training Complete.")
```

## 5. Model Evaluation
We use **Mean Absolute Error (MAE)**.
* This tells us: *"On average, our prediction is off by $X amount."*

```python
# Predict on Test Data
predictions = model.predict(X_test)

# Calculate Error
mae = mean_absolute_error(y_test, predictions)

print(f"Mean Absolute Error: {mae:.2f}")
```

## 6. Real-World Prediction
**Scenario:** The marketing team wants to spend:
* **$150** on TV
* **$25** on Radio
* **$10** on Newspaper

What are the expected sales?

```python
# Client Input
new_campaign = [[150, 25, 10]]

# Generate Prediction
predicted_sales = model.predict(new_campaign)

print(f"Projected Sales: {predicted_sales[0]:.2f} units")
```
